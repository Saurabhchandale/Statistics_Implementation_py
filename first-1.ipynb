```python
#Q 1] 
import random
import math
from collections import Counter

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Mean
def mean(numbers):
    return sum(numbers) / len(numbers)

# Median
def median(numbers):
    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_nums[mid - 1] + sorted_nums[mid]) / 2
    else:
        return sorted_nums[mid]

# (ii) Mode
def mode(numbers):
    counts = Counter(numbers)
    max_count = max(counts.values())
    return [num for num, count in counts.items() if count == max_count]

# (iii) Weighted Mean
def weighted_mean(values, weights):
    return sum(v * w for v, w in zip(values, weights)) / sum(weights)

# (iv) Geometric Mean
def geometric_mean(numbers):
    product = 1
    for num in numbers:
        product *= num
    return product ** (1 / len(numbers))

# (v) Harmonic Mean
def harmonic_mean(numbers):
    return len(numbers) / sum(1 / num for num in numbers)

# (vi) Midrange
def midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# (vii) Trimmed Mean (exclude a percentage of outliers from both ends)
def trimmed_mean(numbers, trim_percent):
    n = len(numbers)
    k = int(n * trim_percent / 100)
    trimmed = sorted(numbers)[k:n - k]
    return mean(trimmed)

# Example usage:
print("Mean:", mean(int_list))
print("Median:", median(int_list))
print("Mode:", mode(int_list))
# For weighted mean, generate random weights
weights = [random.uniform(0.5, 2.0) for _ in range(100)]
print("Weighted Mean:", weighted_mean(int_list, weights))
print("Geometric Mean:", geometric_mean(int_list))
print("Harmonic Mean:", harmonic_mean(int_list))
print("Midrange:", midrange(int_list))
print("Trimmed Mean (10%):", trimmed_mean(int_list, 10))
```

```python
#Q 3]
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not math.isclose(sum(probabilities), 1.0):
            raise ValueError("Probabilities must sum to 1.")
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(v * p for v, p in zip(self.values, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum(((v - mean) ** 2) * p for v, p in zip(self.values, self.probabilities))
```

```python
#Q 4]
import random

# Simulate rolling a fair six-sided die n times
def roll_die(n):
    return [random.randint(1, 6) for _ in range(n)]

# Calculate expected value and variance for a fair six-sided die
def expected_value_die():
    outcomes = [1, 2, 3, 4, 5, 6]
    probabilities = [1/6] * 6
    expected = sum(v * p for v, p in zip(outcomes, probabilities))
    variance = sum(((v - expected) ** 2) * p for v, p in zip(outcomes, probabilities))
    return expected, variance

# Example usage
rolls = roll_die(1000)
expected, variance = expected_value_die()
print("Simulated Rolls (first 10):", rolls[:10])
print("Expected Value:", expected)
print("Variance:", variance)
```

```python
#Q 5]
import numpy as np

def sample_and_stats(distribution, size, **kwargs):
    """
    Generate random samples from a specified distribution and calculate mean and variance.

    Parameters:
        distribution (str): Name of the distribution ('binomial', 'poisson', etc.).
        size (int): Number of samples to generate.
        **kwargs: Parameters for the distribution (e.g., n, p for binomial; lam for poisson).

    Returns:
        samples (np.ndarray): Generated samples.
        mean (float): Mean of samples.
        variance (float): Variance of samples.
    """
    if distribution == 'binomial':
        samples = np.random.binomial(n=kwargs['n'], p=kwargs['p'], size=size)
    elif distribution == 'poisson':
        samples = np.random.poisson(lam=kwargs['lam'], size=size)
    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")
    return samples, np.mean(samples), np.var(samples)

# Example usage:
# bin_samples, bin_mean, bin_var = sample_and_stats('binomial', 1000, n=10, p=0.5)
# pois_samples, pois_mean, pois_var = sample_and_stats('poisson', 1000, lam=3)
```

```python
#Q 7]
import seaborn as sns
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# (i) Function to calculate skewness
def calc_skewness(series):
    """
    Calculate skewness of a pandas Series.
    """
    return series.skew()

# (ii) Program to determine skewness type
def skewness_type(series):
    """
    Determine if the skewness is positive, negative, or approximately symmetric.
    """
    skew = calc_skewness(series)
    if skew > 0.5:
        return "Positive skewness"
    elif skew < -0.5:
        return "Negative skewness"
    else:
        return "Approximately symmetric"

# (iii) Function to calculate covariance
def calc_covariance(series1, series2):
    """
    Calculate covariance between two pandas Series.
    """
    return series1.cov(series2)

# (iv) Function to calculate Pearson correlation coefficient
def calc_pearson_corr(series1, series2):
    """
    Calculate Pearson correlation coefficient between two pandas Series.
    """
    return series1.corr(series2)

# (v) Visualize correlation using scatter plot
def plot_scatter(df, col1, col2):
    """
    Scatter plot to visualize correlation between two columns.
    """
    plt.figure(figsize=(7,5))
    plt.scatter(df[col1], df[col2], alpha=0.7)
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.title(f'Scatter plot of {col1} vs {col2}')
    plt.show()

# Example usage:
print("Skewness of total_bill:", calc_skewness(tips['total_bill']))
print("Skewness of tip:", calc_skewness(tips['tip']))
print("total_bill skewness type:", skewness_type(tips['total_bill']))
print("tip skewness type:", skewness_type(tips['tip']))
print("Covariance between total_bill and tip:", calc_covariance(tips['total_bill'], tips['tip']))
print("Pearson correlation coefficient:", calc_pearson_corr(tips['total_bill'], tips['tip']))
plot_scatter(tips, 'total_bill', 'tip')
```

```python
#Q 8]
def normal_pdf(x, mu, sigma):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
        x (float or np.ndarray): Value(s) at which to evaluate the PDF.
        mu (float): Mean of the distribution.
        sigma (float): Standard deviation of the distribution.

    Returns:
        float or np.ndarray: PDF value(s) at x.
    """
    coef = 1 / (sigma * np.sqrt(2 * np.pi))
    exp_part = np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    return coef * exp_part
```

```python
#Q 9]
def exponential_cdf(x, lam):
    """
    Calculate the cumulative distribution function (CDF) of the exponential distribution.

    Parameters:
        x (float or np.ndarray): Value(s) at which to evaluate the CDF.
        lam (float): Rate parameter (lambda), must be > 0.

    Returns:
        float or np.ndarray: CDF value(s) at x.
    """
    x = np.array(x)
    return 1 - np.exp(-lam * x)
```

```python
#Q 10]
def poisson_pmf(k, lam):
    """
    Calculate the probability mass function (PMF) of the Poisson distribution.

    Parameters:
        k (int or array-like): Number of occurrences.
        lam (float): Expected number of occurrences (lambda).

    Returns:
        float or np.ndarray: PMF value(s) for k.
    """
    k = np.array(k)
    return (np.exp(-lam) * lam**k) / np.math.factorial(k) if k.ndim == 0 else (np.exp(-lam) * lam**k) / np.vectorize(np.math.factorial)(k)
```

```python
#Q 11]
import numpy as np

# 50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)

# 70 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Number of purchases and visitors for each layout
successes = [old_layout.sum(), new_layout.sum()]
nobs = [len(old_layout), len(new_layout)]

# Perform two-proportion z-test
stat, p_value = proportions_ztest(successes, nobs)

print(f"Z-statistic: {stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("The new layout has a significantly higher conversion rate.")
else:
    print("No significant difference in conversion rates between layouts.")
```

```python
#Q 12]
import numpy as np
from statsmodels.stats.weightstats import ztest

# Define example before and after scores for paired z-test
before_program = np.array([70, 75, 80, 65, 78, 72, 74, 77, 69, 73])
after_program = np.array([75, 78, 85, 70, 80, 76, 78, 80, 72, 76])


# Calculate z-test for paired samples (difference in scores)
score_diff = after_program - before_program

# Since we have paired data, test if mean of difference > 0 (one-sided test)
z_stat, p_val = ztest(score_diff, value=0, alternative='larger')

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("The tutoring program significantly improved scores.")
else:
    print("No significant improvement in scores due to the program.")
```

```python
#Q 15]
from scipy import stats
import numpy as np

def ab_test_analysis(layout_a_clicks, layout_b_clicks):
    """
    Perform A/B test analysis using independent two-sample t-test.

    Returns:
        t_stat (float): t-statistic value
        df (float): degrees of freedom
        p_value (float): two-tailed p-value
    """
    # Calculate means and variances
    n1, n2 = len(layout_a_clicks), len(layout_b_clicks)
    mean1, mean2 = np.mean(layout_a_clicks), np.mean(layout_b_clicks)
    var1, var2 = np.var(layout_a_clicks, ddof=1), np.var(layout_b_clicks, ddof=1)

    # Calculate t-statistic
    t_stat = (mean1 - mean2) / np.sqrt(var1/n1 + var2/n2)

    # Degrees of freedom (Welch-Satterthwaite equation)
    df_num = (var1/n1 + var2/n2) ** 2
    df_den = ((var1/n1) ** 2) / (n1 - 1) + ((var2/n2) ** 2) / (n2 - 1)
    df = df_num / df_den

    # Two-tailed p-value
    p_value = 2 * stats.t.sf(np.abs(t_stat), df)

    return t_stat, df, p_value

# Example usage:
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

t_stat, df, p_value = ab_test_analysis(layout_a_clicks, layout_b_clicks)
print(f"t-statistic: {t_stat:.4f}")
print(f"Degrees of freedom: {df:.2f}")
print(f"p-value: {p_value:.4f}")
```

```python
#Q 17]

from scipy import stats

def analyze_intervention(pre_scores, post_scores):
    """
    Perform paired t-test to analyze the impact of an intervention on test scores.

    Parameters:
        pre_scores (list): Scores before intervention.
        post_scores (list): Scores after intervention.

    Returns:
        t_stat (float): t-statistic value.
        p_value (float): two-tailed p-value.
    """
    t_stat, p_value = stats.ttest_rel(post_scores, pre_scores)
    return t_stat, p_value

# Example usage:
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

t_stat, p_value = analyze_intervention(pre_intervention_scores, post_intervention_scores)
print(f"Paired t-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("The intervention had a significant impact on math scores.")
else:
    print("No significant impact of the intervention on math scores.")
```

```python
#Q 19]

from scipy import stats


def compare_quality(version1_scores, version2_scores, alpha=0.05):
    """
    Compare quality scores between two product versions using independent t-test.

    Returns:
        t_stat (float): t-statistic value
        p_value (float): two-tailed p-value
        result (str): Conclusion about significant difference
    """
    t_stat, p_value = stats.ttest_ind(version1_scores, version2_scores, equal_var=False)
    if p_value < alpha:
        result = "There is a significant difference in quality between the two versions."
    else:
        result = "No significant difference in quality between the two versions."
    return t_stat, p_value, result

# Example usage:
# Define example scores for demonstration
version1_scores = [85, 88, 90, 87, 86, 89, 91, 88, 90, 87]
version2_scores = [82, 84, 83, 85, 86, 80, 81, 83, 84, 82]

t_stat, p_value, result = compare_quality(version1_scores, version2_scores)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

```python
#Q 21]

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

np.random.seed(0)

# Correct age group labels
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame
df = pd.DataFrame({
    'Age Group': age_groups,
    'Voter Preference': voter_preferences
})

# Create contingency table
contingency_table = pd.crosstab(df['Age Group'], df['Voter Preference'])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Contingency Table:\n", contingency_table)
print(f"Chi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("There is a significant association between age groups and voter preferences.")
else:
    print("No significant association between age groups and voter preferences.")
```
